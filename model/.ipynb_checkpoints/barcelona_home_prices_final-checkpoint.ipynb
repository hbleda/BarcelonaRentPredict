<h2 style='color:blue'>Outlier Removal Using Standard Deviation and Mean</h2>

<h3 style='color:blue'>Eliminamos outliers basandonos en el precio por m2 y utilizando la desvicación estandar y la media, agrupada por distrito y barrio</h3>

```python
print(df2.shape)
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for (distrito, barrio), subdf in df.groupby(['distrito', 'barrio']):
        m = np.mean(subdf.precio_por_m2)
        st = np.std(subdf.precio_por_m2)
        reduced_df = subdf[(subdf.precio_por_m2>(m-st)) & (subdf.precio_por_m2<=(m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df3 = remove_pps_outliers(df2)
print(df3.shape)
```

```python
import pandas as pd

# Suponiendo que 'df' es tu DataFrame y 'price' es la columna con los precios de alquiler

# Define el número de muestras que deseas para cada rango
desired_samples_1500_2000 = 1000  # Ejemplo
desired_samples_2000_5000 = 2000  # Ejemplo
desired_samples_mas_5000 = 200

# Filtra los datos en los rangos especificados
range_1500_2000 = df5[(df5['precio'] >= 1500) & (df5['precio'] < 2000)]
range_2000_5000 = df5[(df5['precio'] >= 2000) & (df5['precio'] < 5000)]
range_mas_5000 = df5[(df5['precio'] >= 5000)]

# Realiza el submuestreo
subsampled_1500_2000 = range_1500_2000.sample(desired_samples_1500_2000, random_state=42)
subsampled_2000_5000 = range_2000_5000.sample(desired_samples_2000_5000, random_state=42)
subsampled_mas_5000 = range_mas_5000.sample(desired_samples_mas_5000, random_state=42)                     

# Elimina las entradas originales de estos rangos en el DataFrame principal
df5 = df5[~df5.index.isin(range_1500_2000.index)]
df5 = df5[~df5.index.isin(range_2000_5000.index)]
df5 = df5[~df5.index.isin(range_mas_5000.index)]

# Añade las entradas submuestreadas de regreso al DataFrame principal
df5 = pd.concat([df5, subsampled_1500_2000, subsampled_2000_5000,subsampled_mas_5000], axis=0).reset_index(drop=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que 'df' es tu DataFrame y 'price' es la columna con los precios de alquiler
# 1. Visualización de los datos
df5['precio'].hist(bins=50)
plt.xlabel('Precio')
plt.ylabel('Numero de pisos')
plt.title('Distribución de los alquileres')
plt.show()

# 2. Resumen Estadístico
print(df5['precio'].describe())

# 3. Conteo de Precios en Rangos Específicos
bins = [0, 500, 1000, 1500, 2000, 5000, 10000]  # Estos son solo valores de ejemplo, ajústalos según tu contexto
labels = ['<500', '500-1000', '1000-1500', '1500-2000', '2000-5000', '>5000']
df5['precio_range'] = pd.cut(df5['precio'], bins=bins, labels=labels, right=False)
print(df5['precio_range'].value_counts().sort_index())
```

<h2 style='color:blue'>Use K Fold cross validation to measure accuracy of our LinearRegression model</h2>

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

**We can see that in 5 iterations we get a score above 92% all the time. This is pretty good but we want to test few other algorithms for regression to see if we can get even better score. We will use GridSearchCV for this purpose**

<h2 style='color:blue'>Find best model using GridSearchCV</h2>

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False],
                'n_jobs': [-1],
                'positive':[True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
    
find_best_model_using_gridsearchcv(X,y)
```

```python
lr_clf = LinearRegression(fit_intercept=True, positive=False, n_jobs=-1)
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

**Based on above results we can say that LinearRegression gives the best score. Hence we will use that.**

<h2 style='color:blue'>Test the model for few properties</h2>

```python
def predict_price(distrito, barrio, superficie, baños, habitaciones):    
    loc_index_distrito = np.where(X.columns==distrito)[0][0]
    loc_index_barrio = np.where(X.columns==barrio)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = superficie
    x[1] = baños
    x[2] = habitaciones
    
    if loc_index_distrito >= 0:
        x[loc_index_distrito] = 1
    if loc_index_barrio >= 0:
        x[loc_index_barrio] = 1

    return lr_clf.predict([x])[0]
```

```python
predict_price('la Font d\'en Fargues','Horta-Guinardo',17, 0, 0)
```

```python
predict_price('Sant Andreu','Navas',80, 1, 3)
```

```python
predict_price('Sant Martí','el Clot',44, 1, 2)
```
