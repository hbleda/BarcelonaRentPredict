<h2 style='color:blue'>Feature Engineering</h2>

<h2 style='color:blue'>Outlier Removal Using Standard Deviation and Mean</h2>

<h3 style='color:blue'>Eliminamos outliers basandonos en el precio por m2 y utilizando la desvicación estandar y la media, agrupada por distrito y barrio</h3>

```python
print(df2.shape)
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for (distrito, barrio), subdf in df.groupby(['distrito', 'barrio']):
        m = np.mean(subdf.precio_por_m2)
        st = np.std(subdf.precio_por_m2)
        reduced_df = subdf[(subdf.precio_por_m2>(m-st)) & (subdf.precio_por_m2<=(m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df3 = remove_pps_outliers(df2)
print(df3.shape)
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
plt.hist(df3.precio_por_m2,rwidth=0.8)
plt.xlabel("Precio por m2")
plt.ylabel("Cuenta")
```

<h2 style='color:blue'>Use One Hot Encoding For distrito</h2>

```python
dummies_barrio = pd.get_dummies(df5.barrio)
dummies_barrio.head(10)
```

```python
dummies_distrito = pd.get_dummies(df5.distrito)
dummies_distrito.head(10)
```

```python
df6 = pd.concat([df5,dummies_barrio,dummies_distrito],axis='columns')
df6.head()
```

```python
df7 = df6.drop('barrio',axis='columns')
df8 = df7.drop('distrito',axis='columns')
df9 = df8.drop('precio_range',axis='columns')
df9.head(10)
```

<h2 style='color:blue'>Ahora ya podemos construir nuestro modelo...</h2>

```python
df9.shape
```

```python
X = df9.drop(['precio'],axis='columns')
X.head(10)
```

```python
X.shape
```

```python
y = df9.precio
y.head(10)
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

<h2 style='color:blue'>Use K Fold cross validation to measure accuracy of our LinearRegression model</h2>

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

**We can see that in 5 iterations we get a score above 86% all the time. This is pretty good but we want to test few other algorithms for regression to see if we can get even better score. We will use GridSearchCV for this purpose**

<h2 style='color:blue'>Find best model using GridSearchCV</h2>

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
pd.set_option('display.max_colwidth', None)

def find_best_model_using_gridsearchcv(X, y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False],
                'n_jobs': [-1],
                'positive': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
                'splitter': ['best', 'random']
            }
        },
        'svr': {
            'model': SVR(),
            'params': {
                'C': [0.01, 0.1, 1, 10],
                'epsilon': [0.01, 0.1, 1],
                'kernel': ['linear', 'rbf']
            }
        },
        'random_forest': {
            'model': RandomForestRegressor(),
            'params': {
                'n_estimators': [10, 50, 100],
                'max_depth': [None, 10, 20],
                'min_samples_split': [2, 5, 10],
                'min_samples_leaf': [1, 2, 4]
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['best_params','model', 'best_score'])

# Suponiendo que X y y ya están definidos anteriormente
find_best_model_using_gridsearchcv(X, y)
```

```python
print(find_best_model_using_gridsearchcv(X, y).iloc[4]['best_params'])
```

```python
rf_clf = RandomForestRegressor(max_depth=None, min_samples_leaf=4, min_samples_split=5,n_estimators=100)
rf_clf.fit(X_train,y_train)
rf_clf.score(X_test,y_test)
```

**Based on above results we can say that LinearRegression gives the best score. Hence we will use that.**

<h2 style='color:blue'>Test the model for few properties</h2>

```python
def predict_price(distrito, barrio, superficie, baños, habitaciones):    
    loc_index_distrito = np.where(X.columns==distrito)[0][0]
    loc_index_barrio = np.where(X.columns==barrio)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = superficie
    x[1] = baños
    x[2] = habitaciones
    
    if loc_index_distrito >= 0:
        x[loc_index_distrito] = 1
    if loc_index_barrio >= 0:
        x[loc_index_barrio] = 1

    return rf_clf.predict([x])[0]
```

```python
predict_price('Horta-Guinardó','la Font d\'en Fargues',17, 1, 2)
```

```python
predict_price('Sant Andreu','Navas',80, 1, 3)
```

```python
predict_price('Sant Martí','el Clot',44, 1, 2)
```
